```python
features = ['price','dlc_count','metacritic_score','recommendations','peak_ccu','num_reviews_total','avg_estimated_owners','num_reviews_recent']

X = df[features]
y = df['cat_pos_totals_encoded']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=42)
X_train, y_train = ros.fit_resample(X_train, y_train)


model = GradientBoostingClassifier()
model.fit(X_train, y_train)
```

```python
X_train
```

```python
lr_model = LogisticRegression(max_iter=1000)
rf_model = RandomForestClassifier()
gb_model = GradientBoostingClassifier()
gbr_model = GradientBoostingRegressor()
```

```python
lr_model.fit(X_train, y_train)
rf_model.fit(X_train, y_train)
gb_model.fit(X_train, y_train)
gbr_model.fit(X_train, y_train)
```

```python
def evaluate_model(model,X_test,y_test,name='Model'):
    y_pred = model.predict(X_test)
    if name.lower().find('regress') != -1:
        y_pred = np.clip(np.round(y_pred), 0, 2).astype(int)
    acc = accuracy_score(y_test, y_pred)

    print(f'-- {name} --')
    print(f'Accuracy: {acc:.2f}')
    print(classification_report(y_test, y_pred))
    print(confusion_matrix(y_test,y_pred))
    print('-'*30)
```

```python
y.value_counts()
```

```python
evaluate_model(lr_model, X_test, y_test, 'LogisticRegression')
evaluate_model(rf_model, X_test, y_test, 'RandomForestClassifier')
evaluate_model(gb_model, X_test, y_test, 'GradientBoostingClassifier')
evaluate_model(gbr_model, X_test, y_test, 'GradientBoostingRegression')
```

```python
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_auc_score

y_true = y_test  

y_true_bin = label_binarize(y_true, classes=[0,1,2])

y_pred_proba = rf_model.predict_proba(X_test)

roc_auc = roc_auc_score(y_true_bin, y_pred_proba, multi_class='ovr')
print(f'Multi-class ROC AUC: {roc_auc:.4f}')

```

```python
def predict_games(price,dlc_count,metacritic_score,recommendations,peak_ccu,num_reviews_total,avg_estimated_owners,num_reviews_recent):
    model = joblib.load('model.joblib')
    input_arr = np.array([[price,dlc_count,metacritic_score,recommendations,peak_ccu,num_reviews_total,avg_estimated_owners,num_reviews_recent]])
    predict = model.predict(input_arr)
    return predict.round(2)[0]
```
