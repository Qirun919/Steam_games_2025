```python
# all the thing inside here is number 
features = [
    "price",
    "required_age",
    "dlc_count",
    "metacritic_score",
    "achievements",
    "recommendations",
    "discount",
    "peak_ccu",
    "windows",
    "mac",
    "linux"
]

target = "average_playtime_forever"

max_playtime_threshold = 1e5  # 1e5 = 1 Ã— 10^5 = 100,000

# need to clean it again or it will be minus
df[features] = df[features].fillna(0)

model_cols = features + [target] # only need this
df_model = df[model_cols]
df_model = df_model[df_model[target] < max_playtime_threshold]
df_model = df_model[df_model[target] > 0]

X = df_model.drop(target, axis=1)
y = df_model[target]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.dtypes
```

```python
lr_model = LogisticRegression()
rfc_model = RandomForestClassifier()
gbc_model = GradientBoostingClassifier()
```

```python
# Logistic Regression
lr_model.fit(X_train, y_train)
y_pred = lr_model.predict(X_test)
y_proba = lr_model.predict_proba(X_test)[:,1]

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d")
plt.title("Logistic Regression Confusion Matrix")
plt.show()
```
